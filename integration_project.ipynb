```python
from google.colab import files
import pandas as pd
import io

print("Please upload your 'augmented_sonar_dataset.csv' file.")
uploaded = files.upload()

# Check if the correct file was uploaded
if 'augmented_sonar_dataset.csv' in uploaded:
    # Read the uploaded file directly into the DataFrame
    df = pd.read_csv(io.BytesIO(uploaded['augmented_sonar_dataset.csv']))

    print("\n✅ Dataset loaded successfully!")
    print("Here is a preview of your data:")
    display(df.head())
else:
    print("\n❌ Error: The file 'augmented_sonar_dataset.csv' was not uploaded.")
    print("Please make sure the file you select is named correctly and try again.")
```

```python
# --- 2. Define parameters and generate location data ---

# Set the size of your tactical grid
grid_size = 10
# Create an empty list to store the new location data
locations = []
# Identify the 60 sonar feature columns
sonar_features = [f'Sonar_{i}' for i in range(1, 61)]


print("⚙️  Generating threat locations based on strongest sonar signal...")

# Loop through each row (each event) in your dataset
for index, row in df.iterrows():
    # Only generate a location if it's a confirmed threat (Threat == 1)
    if row['Threat'] == 1:
        # Find which sonar sensor gave the strongest reading
        strongest_sensor_index = np.argmax(row[sonar_features].values)

        # Convert that sensor index (0-59) into an angle (0-360 degrees)
        angle = (strongest_sensor_index / 59.0) * 360.0
        rad = np.deg2rad(angle)

        # Calculate an (X, Y) coordinate on the grid based on the angle
        center = grid_size / 2
        radius = center - 1
        x = int(center + radius * np.cos(rad))
        y = int(center + radius * np.sin(rad))

        # Make sure the coordinate is within the grid boundaries (0 to 9)
        x = max(0, min(grid_size - 1, x))
        y = max(0, min(grid_size - 1, y))

        locations.append((x, y))
    else:
        # If it's not a threat, there is no location, so we add 'None'
        locations.append((None, None))

print(f"\n✅ Location data generated for {len(locations)} events.")
```

```python
# --- 3. Add the new columns and save the file ---

# Add the list of locations as two new columns in your DataFrame
df[['threat_x', 'threat_y']] = locations

# Save this new, complete dataset to a new CSV file
df.to_csv("master_dataset.csv", index=False)

print("✅ 'master_dataset.csv' has been created successfully!")
print("Here's a final preview of your new master dataset with the location columns:")

# Show the Threat column and the new location columns for the first 10 rows
display(df[['Threat', 'threat_x', 'threat_y']].head(10))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# --- 4. Load the master dataset and prepare for modeling ---
df_master = pd.read_csv("master_dataset.csv")

# Define our features (the 63 sensor readings)
features = [f'Sonar_{i}' for i in range(1, 61)] + ['Radar_Signal', 'Infrared', 'Speed']
X = df_master[features]

# Define our three separate targets
y_threat = df_master['Threat']
y_x = df_master['threat_x']
y_y = df_master['threat_y']

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_threat_train, y_threat_test, y_x_train, y_x_test, y_y_train, y_y_test = train_test_split(
    X, y_threat, y_x, y_y, test_size=0.3, random_state=42, stratify=y_threat
)

print("✅ Data successfully split into training and testing sets.")
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# --- 5. Train and Compare Threat Classifiers ---

# First, we need to scale our features. This is very important for SVM.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("⚙️  Training Naive Bayes, Random Forest, and SVM models...")
print("-" * 30)

# Initialize the models
models = {
    "Naive Bayes": GaussianNB(),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM": SVC(kernel='rbf', probability=True, random_state=42)
}

# Train and evaluate each model
for name, model in models.items():
    # Train the model
    model.fit(X_train_scaled, y_threat_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test_scaled)

    # Calculate and print the accuracy
    acc = accuracy_score(y_threat_test, y_pred)
    print(f"✅ {name} Accuracy: {acc:.3f}")

# For now, let's assume SVM is our best model for the next steps
best_threat_model = models["SVM"]
```

```python
from sklearn.ensemble import RandomForestClassifier

# --- 6. Train the Location Predictors ---

# Combine the training features and location targets into one DataFrame
# This makes it easy to filter out the non-threat samples
train_df = pd.concat([X_train.reset_index(drop=True),
                      y_x_train.reset_index(drop=True),
                      y_y_train.reset_index(drop=True)], axis=1)

# Drop all rows where the location is NaN (i.e., where there was no threat)
loc_train_df = train_df.dropna()

print(f"Original training samples: {len(train_df)}")
print(f"Training samples with threat locations: {len(loc_train_df)}")

# Separate the filtered data back into features (X) and targets (y)
X_train_loc = loc_train_df[features]
y_x_train_loc = loc_train_df['threat_x']
y_y_train_loc = loc_train_df['threat_y']

# --- Train the X and Y models ---
print("\n⚙️  Training Location Predictor models...")

# Initialize and train the model for the X-coordinate
loc_x_model = RandomForestClassifier(n_estimators=100, random_state=42)
loc_x_model.fit(X_train_loc, y_x_train_loc)

# Initialize and train the model for the Y-coordinate
loc_y_model = RandomForestClassifier(n_estimators=100, random_state=42)
loc_y_model.fit(X_train_loc, y_y_train_loc)

print("✅ Location X and Y models have been trained successfully.")
```

```python
import joblib

# --- 7. Save all trained models and the scaler to files ---

# Save the "Gatekeeper" SVM model
joblib.dump(best_threat_model, 'threat_model.pkl')

# Save the "Navigator" location models
joblib.dump(loc_x_model, 'loc_x_model.pkl')
joblib.dump(loc_y_model, 'loc_y_model.pkl')

# Save the scaler
joblib.dump(scaler, 'scaler.pkl')

print("✅ All models and the scaler have been saved successfully!")
print("You should now see the following files in your Colab file pane:")
print("- threat_model.pkl")
print("- loc_x_model.pkl")
print("- loc_y_model.pkl")
print("- scaler.pkl")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
import time

# --- 1. Page Configuration ---
st.set_page_config(
    page_title="Smart Escapism - Command Center",
    layout="wide",
    page_icon="🛰️"
)

st.title("⚓ Smart Escapism: AI Command Center")

# --- 2. Load Data ---
@st.cache_data
def load_data():
    try:
        df = pd.read_csv("master_dataset.csv")
        return df
    except FileNotFoundError:
        st.error("Error: `master_dataset.csv` not found. Please upload it to your Colab session.")
        return None

df = load_data()

if df is not None:
    # --- 3. Sidebar Controls ---
    st.sidebar.header("Mission Configuration")
    threat_model_choice = st.sidebar.selectbox(
        "Choose Threat Classifier (Gatekeeper)",
        ("SVM", "Random Forest", "Naive Bayes")
    )
    location_model_choice = st.sidebar.selectbox(
        "Choose Location Predictor (Navigator)",
        ("Random Forest",) # Add more if you want later
    )
    train_button = st.sidebar.button("Train Models & Evaluate", type="primary")

    # --- 4. Training Pipeline ---
    def train_pipeline(data, threat_model_name, loc_model_name):
        # Prepare data
        features = [f'Sonar_{i}' for i in range(1, 61)] + ['Radar_Signal', 'Infrared', 'Speed']
        X = data[features]
        y_threat = data['Threat']
        y_x = data['threat_x']
        y_y = data['threat_y']

        # Split data
        X_train, X_test, y_threat_train, y_threat_test, y_x_train, y_x_test, y_y_train, y_y_test = train_test_split(
            X, y_threat, y_x, y_y, test_size=0.3, random_state=42, stratify=y_threat
        )

        # Scale features
        scaler = StandardScaler().fit(X_train)
        X_train_scaled = scaler.transform(X_train)
        X_test_scaled = scaler.transform(X_test)

        # --- Train Threat Model ---
        if threat_model_name == "SVM":
            threat_model = SVC(probability=True, random_state=42)
        elif threat_model_name == "Random Forest":
            threat_model = RandomForestClassifier(random_state=42)
        else:
            threat_model = GaussianNB()
        threat_model.fit(X_train_scaled, y_threat_train)

        # --- Train Location Models (on threat data only) ---
        train_df = pd.concat([X_train.reset_index(drop=True), y_x_train.reset_index(drop=True), y_y_train.reset_index(drop=True)], axis=1)
        loc_train_df = train_df.dropna()
        X_train_loc = loc_train_df[features]

        loc_x_model = RandomForestClassifier(random_state=42).fit(X_train_loc, loc_train_df['threat_x'])
        loc_y_model = RandomForestClassifier(random_state=42).fit(X_train_loc, loc_train_df['threat_y'])

        # --- Evaluate Threat Model ---
        y_pred = threat_model.predict(X_test_scaled)
        y_prob = threat_model.predict_proba(X_test_scaled)[:, 1]
        accuracy = accuracy_score(y_threat_test, y_pred)
        cm = confusion_matrix(y_threat_test, y_pred)
        fpr, tpr, _ = roc_curve(y_threat_test, y_prob)
        roc_auc = auc(fpr, tpr)

        return {
            "threat_model": threat_model, "loc_x_model": loc_x_model, "loc_y_model": loc_y_model, "scaler": scaler,
            "accuracy": accuracy, "cm": cm, "fpr": fpr, "tpr": tpr, "roc_auc": roc_auc, "X_test": X_test
        }

    # --- 5. Main App Logic ---
    if train_button:
        with st.spinner("Training AI models... This may take a moment."):
            results = train_pipeline(df, threat_model_choice, location_model_choice)
            st.session_state['results'] = results # Save results to session
        st.success("Training Complete!")

    if 'results' in st.session_state:
        results = st.session_state['results']

        st.header("1. Threat Model Evaluation")
        st.metric("Model Accuracy", f"{results['accuracy']:.3f}")

        col1, col2 = st.columns(2)
        with col1:
            st.subheader("Confusion Matrix")
            fig, ax = plt.subplots()
            sns.heatmap(results['cm'], annot=True, fmt='d', cmap='Blues', ax=ax,
                        xticklabels=['Safe', 'Threat'], yticklabels=['Safe', 'Threat'])
            ax.set_xlabel("Predicted Label")
            ax.set_ylabel("True Label")
            st.pyplot(fig)
        with col2:
            st.subheader("ROC Curve")
            fig2, ax2 = plt.subplots()
            ax2.plot(results['fpr'], results['tpr'], lw=2, label=f"AUC = {results['roc_auc']:.3f}")
            ax2.plot([0, 1], [0, 1], 'r--')
            ax2.set_xlabel("False Positive Rate")
            ax2.set_ylabel("True Positive Rate")
            ax2.legend()
            st.pyplot(fig2)

        st.header("2. Live Decoy Deployment Simulation")
        threshold = st.slider("Set Decoy Deployment Threshold", 0.0, 1.0, 0.6, 0.01)

        if st.button("Run Simulation"):
            st.subheader("Simulation Map")
            asset_loc = (2, 2)
            map_placeholder = st.empty()
            log_placeholder = st.empty()

            for i in range(len(results['X_test'])):
                sample_df = results['X_test'].iloc[i:i+1]
                sample_scaled = results['scaler'].transform(sample_df)

                # Predict threat
                prob = results['threat_model'].predict_proba(sample_scaled)[0][1]

                # Create grid for map
                fig3, ax3 = plt.subplots()
                ax3.set_xticks(np.arange(10))
                ax3.set_yticks(np.arange(10))
                ax3.grid()
                ax3.scatter(asset_loc[0], asset_loc[1], c='lime', marker='s', s=200, label='Asset')

                if prob >= threshold:
                    # Predict location
                    pred_x = results['loc_x_model'].predict(sample_df)[0]
                    pred_y = results['loc_y_model'].predict(sample_df)[0]
                    # Calculate decoy
                    decoy_x = (asset_loc[0] + pred_x) // 2
                    decoy_y = (asset_loc[1] + pred_y) // 2

                    ax3.scatter(pred_x, pred_y, c='red', marker='X', s=200, label='Threat')
                    ax3.scatter(decoy_x, decoy_y, c='cyan', marker='o', s=200, label='Decoy')
                    log_text = f"Cycle {i+1}: Threat probability {prob:.2f} -> **DEPLOYING DECOY** at ({decoy_x}, {decoy_y})"
                else:
                    log_text = f"Cycle {i+1}: Threat probability {prob:.2f} -> Monitoring..."

                ax3.legend()
                map_placeholder.pyplot(fig3)
                log_placeholder.info(log_text)
                time.sleep(1)
                plt.close(fig3)
```
